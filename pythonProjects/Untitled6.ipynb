```python
import numpy as np

# Generate some data
data = np.random.normal(0, 1, 1000)

# Plot histogram with density=True
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot a normal distribution for comparison
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 175
std_dev = 10

# Generate random data
data = np.random.normal(mean, std_dev, 1000)

# Plot
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-0.5 * ((x - mean) / std_dev)**2) / (std_dev * np.sqrt(2 * np.pi))
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 175
std_dev = 10

# Generate random data
data = np.random.normal(mean, std_dev, 1000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Plot the normal distribution curve using scipy's norm.pdf
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 175
std_dev = 10

# Generate random data
data = np.random.normal(mean, std_dev, 1000)

# Plot histogram with automatic bin calculation
plt.hist(data, bins='auto', density=True, alpha=0.6, color='g')

# Plot the normal distribution curve using scipy's norm.pdf
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10    # number of trials
p = 0.5   # probability of success

# Generate random data
data = np.random.binomial(n, p, 1000)

# Plot
plt.hist(data, bins=np.arange(n+2)-0.5, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the binomial distribution PMF
x = np.arange(n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'o', color='red')
plt.title('Binomial Distribution')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter
lam = 3  # average rate (λ)

# Generate random data
data = np.random.poisson(lam, 1000)

# Plot
n = max(data)
plt.hist(data, bins=np.arange(n + 2) - 0.5, density=True, alpha=0.6, color='r', edgecolor='black')

# Plot the Poisson distribution PMF
x = np.arange(n + 1)
pmf = poisson.pmf(x, lam)
plt.plot(x, pmf, 'o', color='black')
plt.title('Poisson Distribution')
plt.xlabel('Number of Emails per Hour')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Read data from CSV
df = pd.read_csv(r'C:\Users\DELL\Downloads\student.csv')
data = df['emails_per_hour'].to_numpy()

# Parameter
lam = np.mean(data)  # Calculate the average rate (λ) from the data

# Plot
n = max(data)
plt.hist(data, bins=np.arange(n + 2) - 0.5, density=True, alpha=0.6, color='r', edgecolor='black')

# Plot the Poisson distribution PMF
x = np.arange(n + 1)
pmf = poisson.pmf(x, lam)
plt.plot(x, pmf, 'o', color='black')
plt.title('Poisson Distribution')
plt.xlabel('Number of Emails per Hour')
plt.ylabel('Probability')
plt.show()
```
