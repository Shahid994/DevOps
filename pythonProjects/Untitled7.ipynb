```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate data with positive association
np.random.seed(0)
x_pos = np.random.rand(100)
y_pos = 2 * x_pos + np.random.normal(0, 0.1, 100)  # y is positively associated with x

# Plot the data
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_pos, y=y_pos)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Positive Association: Scatter Plot of X vs Y')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation_pos = np.corrcoef(x_pos, y_pos)[0, 1]
print(f'Positive Association Correlation Coefficient: {correlation_pos:.2f}')
```

```python
# Generate data with negative association
x_neg = np.random.rand(100)
y_neg = -2 * x_neg + np.random.normal(0, 0.1, 100)  # y is negatively associated with x

# Plot the data
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_neg, y=y_neg)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Negative Association: Scatter Plot of X vs Y')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation_neg = np.corrcoef(x_neg, y_neg)[0, 1]
print(f'Negative Association Correlation Coefficient: {correlation_neg:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values and their corresponding probabilities
values = np.array([1, 2, 3])
probabilities = np.array([0.2, 0.5, 0.3])

# Calculate the CDF
cdf = np.cumsum(probabilities)

# Plot the PMF
plt.stem(values, probabilities, basefmt=" ")
plt.xlabel('Value')
plt.ylabel('PMF')
plt.title('Probability Mass Function (PMF)')
plt.show()

# Plot the CDF
plt.step(values, cdf, where='post', label='CDF')
plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.show()
```
